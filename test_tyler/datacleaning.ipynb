```python
import pandas as pd
from pathlib import Path
```

```python
dataset = pd.read_csv("../test_emily/MH_merged_cleaned.csv")
dataset.head()
```

```python
mergewith = pd.read_csv("./countries.csv")
mergewith.head()
mergewith.rename(columns={"name":"Entity"}, inplace=True)
```

```python
mergewith.info()
```

```python
merged = pd.merge(dataset, mergewith, on="Entity")
merged.head()
```

```python
merged.drop(columns=["DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Depressive disorders", "DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Schizophrenia", "DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Bipolar disorder", "DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Eating disorders", "DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Anxiety disorders", "country"], inplace=True)
merged.head()
```

```python
merged.rename(columns={"Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized":"Schizophrenia Rate", "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized":"Depression Rate", "Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized":"Anxiety Rate", "Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized":"Bipolar Rate", "Eating disorders (share of population) - Sex: Both - Age: Age-standardized":"ED Rate"}, inplace=True)
merged.head()
```

```python
merged2 = merged.loc[merged["Year"] == 2019]
merged2.head()
```

```python
filepath = Path("mapdata.csv")
filepath.parent.mkdir(parents=True, exist_ok=True)
merged2.to_csv(filepath)
```
