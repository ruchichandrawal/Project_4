```python
import pandas as pd
```

```python
df_1 = pd.read_csv("1- mental-illnesses-prevalence.csv")
df_1
```

```python
df_2 = pd.read_csv("2- burden-disease-from-each-mental-illness(1).csv")
df_2
```

```python
combo = df_1.merge(df_2, how= "inner", on=["Entity","Code","Year"])
combo
```

```python
combo = combo.loc[combo["Year"]==2014]
combo
```

```python
combo = combo.drop(index = 6054)
combo
```

```python
combo.rename(columns={"Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized": "Schizophrenia Prevelance", 
"Depressive disorders (share of population) - Sex: Both - Age: Age-standardized": "Depression Prevalence", 
"Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized" : "Anxiety Prevelance", 
"Eating disorders (share of population) - Sex: Both - Age: Age-standardized": "Eating Disorder Prevalence", 
"Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized": "Bipolar Prevalence",
"DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Depressive disorders" : "DALYs- Depression", 
"DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Schizophrenia": "DALYs- Schizophrenia", 
"DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Bipolar disorder" : "DALYs- Bipolar",
"DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Eating disorders" : "DALYs- Eating Disorder",
"DALYs (rate) - Sex: Both - Age: Age-standardized - Cause: Anxiety disorders" : "DALYs- Anxiety"

},inplace = True)
```

```python
combo.isna().count()
```

```python
combo["Overall Prevalence"] = "NONE"

combo.loc[combo["Anxiety Prevelance"] == max(combo["Anxiety Prevelance"]), "Overall Prevalence"] = "MAX ANXIETY"
combo.loc[combo["Depression Prevalence"] == max(combo["Depression Prevalence"]), "Overall Prevalence"] = "MAX DEPRESSION"
combo.loc[combo["Bipolar Prevalence"] == max(combo["Bipolar Prevalence"]), "Overall Prevalence"] = "MAX BIPOLAR"
combo.loc[combo["Eating Disorder Prevalence"] == max(combo["Eating Disorder Prevalence"]), "Overall Prevalence"] = "MAX EATING DISORDER"
combo.loc[combo["Schizophrenia Prevelance"] == max(combo["Schizophrenia Prevelance"]), "Overall Prevalence"] = "MAX SCHIZOPHRENIA"
combo.loc[combo["Overall Prevalence"] != "NONE"]
```

```python
#combo["Overall Prevalence"] = "NONE"

combo.loc[combo["Anxiety Prevelance"] == min(combo["Anxiety Prevelance"]), "Overall Prevalence"] = "MIN ANXIETY"
combo.loc[combo["Depression Prevalence"] == min(combo["Depression Prevalence"]), "Overall Prevalence"] = "MIN DEPRESSION"
combo.loc[combo["Bipolar Prevalence"] == min(combo["Bipolar Prevalence"]), "Overall Prevalence"] = "MIN BIPOLAR"
combo.loc[combo["Eating Disorder Prevalence"] == min(combo["Eating Disorder Prevalence"]), "Overall Prevalence"] = "MIN EATING DISORDER"
combo.loc[combo["Schizophrenia Prevelance"] == min(combo["Schizophrenia Prevelance"]), "Overall Prevalence"] = "MIN SCHIZOPHRENIA"
combo.loc[combo["Entity"] == "Somalia", "Overall Prevalence"] = "MIN EATING DISORDER & MIN SCHIZOPHRENIA"
combo.loc[combo["Overall Prevalence"] != "NONE"]
```

```python
combo.to_csv("tableau.csv", index = False)
```
